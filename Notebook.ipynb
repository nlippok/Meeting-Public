<h2><u>How about Widgets?</u></h2>

Here is CR imaging using FC-FDML ...

```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
from ipywidgets import widgets
import numpy as np

img = mpimg.imread('im2.png') 
plt.imshow(img, cmap=plt.cm.Greys) 
```

Here I apply a simple horizontal LPF to the image ...

```python
def Filt(ind):
    Img = np.fft.ifft2(img)
    Img[:,ind:-1,:] = 0
    img2 = np.fft.fft2(Img)
    plt.imshow(np.real(img2), cmap=plt.cm.Greys) 
    
ind_slide = widgets.IntSlider(value = 50,
                              description = 'Filter',
                              min = 1,
                              max = 50,
                              step = 5,
                              continuous_update = True)

widgets.interact(Filt, ind = ind_slide)
```
